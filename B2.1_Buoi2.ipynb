```python
# Khai bao thư viẹn import matplotlib.pyplot as plt Import "matplotlib.pyplot
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model, metrics
import pandas as pd # Đọc dữ liệu từ file csv
df = pd.read_csv("ds_n.csv", index_col = 0)
print(df)
# Vẽ biểu đồ minh họa dataset để xem sự phân bố của dữ liệu
plt.plot(df.ChieuCao, df.CanNang, 'ro')
plt.xlabel( 'Chiều cao (cm)')
plt.ylabel( 'Cân nặng (kg)')
plt.show() #Removed extra indentation here
```

```python
x = df.loc[:, ['ChieuCao']].values
y = df.CanNang.values
model = linear_model.LinearRegression()
model.fit(x, y)
y_predict = model.predict(x) # Changed X to x to match the training data
# Độ đo
mse = metrics.mean_squared_error(y_predict, y)
r_2 = metrics.r2_score(y, model.predict(x)) # Changed X to x
print("Tổng bình phương sai số trên tập mẫu:", mse)
print("Hệ số xác định R2:", r_2)
print("Hệ số hồi quy:", model.coef_)
print("Sai số:", model.intercept_)
print(f"Công thức: [Nặng] = {model.coef_[0]} x [Cao] + {model.intercept_}")
```

```python
# Vẽ đường hôi quy so với tập dữ liệu
plt.scatter(x, y, color='red') # Changed X to x for consistency
plt.plot(x, model.predict(x), color='green') # Changed X to x
plt.title('Linear Regression')
plt.xlabel('Chiều cao')
plt.ylabel('Cân nặng') # Corrected plt-ylabel to plt.ylabel
plt.show() # Removed extra indentation
```
